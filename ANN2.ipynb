```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
df
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
```

```python
df['Geography']=df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(6,activation='sigmoid'))
```

```python
ann.add(tf.keras.layers.Dense(1,activation='relu'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50)
```

```python
pred=ann.predict(scaler.transform([x_train[6]]))
if(pred>0.5):
  print(1)
else:
  print(0)
```
